# **EDA on Sleep Health Lifestyle Dataset**

**Explorative Data Analysis (EDA)**

This dataset explores the relationships between sleep, lifestyle habits, and health outcomes. Key components include:

*   Sleep Data: Duration, quality, and potential sleep disorders.

*   Lifestyle Metrics: Physical activity level, daily steps, and stress level.

*   Health Metrics: BMI category, heart rate, blood pressure.

```python
import pandas as pd
import matplotlib as plt
import seaborn as sns
```

```python
data = pd.read_csv('/content/sleep_health_lifestyle_dataset.csv')
data
```

**Potential Study Objectives**


*   Analyze Sleep Patterns:
    *   Explore distributions of sleep duration and quality.
    *   Identify factors (e.g., age, gender, stress) affecting sleep quality.


*   Correlate Lifestyle with Health:
    *   Assess the impact of physical activity on BMI, heart rate, and sleep quality.

    *   Investigate if higher stress levels are associated with poorer sleep and health outcomes.


*   Identify Health Risks:
    *   Examine relationships between BMI, blood pressure, and heart rate.

    *   Determine if individuals with low sleep quality have higher stress or health risks.





```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme(style="whitegrid")

numeric_columns = data.select_dtypes(include=["int64", "float64"]).columns

plt.figure(figsize=(12, 8))
for i, col in enumerate(numeric_columns, 1):
    plt.subplot(3, 3, i)
    sns.histplot(data[col], kde=True, color="skyblue")
    plt.title(f"Distribution of {col}")
plt.tight_layout()
plt.show()
```

**Statistical and Visual Techniques**

*   Univariate Analysis:
    *   Histograms and boxplots for distributions of sleep duration, stress levels, etc.



*   Bivariate Analysis:
    *   Scatter plots and correlation analysis for numerical relationships.
    *   Bar plots to analyze categorical relationships (e.g., occupation vs. sleep quality).


*   Multivariate Analysis:

    *   Heatmaps to identify clusters of correlated variables.
    *   Pairplots for visually exploring feature relationships.


*   Group Comparisons:
    *   Compare health metrics across groups like gender or BMI categories using group-wise means.












```python
corr_matrix = data[numeric_columns].corr()

plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, fmt=".2f", cmap="coolwarm", cbar=True)
plt.title("Correlation Matrix")
plt.show()
```
