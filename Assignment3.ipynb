```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

1. Basic Line Plot

for ùë• values ranging from -10 to 10.
o Label the x-axis as "X Values" and the y-axis as "Squared Values."

```python
x=np.arange(-10,11)
y=x**2
plt.plot(x,y)
plt.xlabel("X Values")
plt.ylabel("Squared Values")
plt.title("Squared Values")
plt.show()
```

4. Histogram
o Simulate 100 random values from a normal distribution using
np.random.randn() and plot a histogram.
o Customize the number of bins and add a grid to the plot.


```python
data = np.random.randn(100)
plt.hist(data, bins=10, edgecolor='black')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histogram of Random Values')
plt.grid(True)
plt.show()
```

5. Pie Chart
o Visualize the percentage distribution of five categories ['A', 'B', 'C', 'D', 'E']
with values [20, 30, 25, 15, 10].
o Display the percentage values on the chart and use a color palette for
differentiation.


```python
categories = ['A', 'B', 'C', 'D', 'E']
values = [20, 30, 25, 15, 10]
colors = ['blue', 'yellow', 'red', 'green', 'white']
plt.pie(values, labels=categories, colors=colors, autopct='%1.1f%%')
plt.title('percentage Distribution of categories')
plt.axis('equal')
plt.show()
```

7. Heatmap (Seaborn)
o Create a 5x5 matrix with random integers between 1 and 10 using
np.random.randint() and visualize it as a heatmap using Seaborn.
o Add a color bar and annotate the heatmap with the values.


```python
data = np.random.randint(1, 11, size=(5, 5))
plt.figure(figsize=(8, 6))
plt.imshow(data, cmap='hot', interpolation='nearest')
plt.colorbar(label='Value')
plt.xticks(range(5), range(1, 6))
plt.yticks(range(5), range(1, 6))
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Heatmap of Random Integers')
plt.show()
```

9. Pair Plot (Seaborn)
o Manually create a small dataset with columns such as 'Age', 'Height',
'Weight' for 10 individuals and visualize the relationships using a pair plot.


```python
import seaborn as sns
data = {'Age': [25, 30, 22, 28, 35, 20, 27, 32, 29, 24],
        'Height': [175, 180, 165, 172, 185, 160, 170, 182, 178, 168],
        'Weight': [70, 75, 60, 68, 80, 55, 65, 78, 72, 62]}
df = pd.DataFrame(data)
sns.pairplot(df)
plt.suptitle("Pair Plot of Age, Height, and Weight", y=1.02)
plt.show()
```

10. Save and Style Plots
